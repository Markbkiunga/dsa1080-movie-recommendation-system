```python
# Week 1 - Traffic Accident Severity Project: Data Loading & Initial Overview

import pandas as pd

# Load datasets
clean_df = pd.read_csv('../data/cleaned.csv')
raw_df = pd.read_csv('../data/rta.csv')

print("✅ Datasets Loaded Successfully")

# --- Basic Info ---
print("\n🔍 Cleaned Dataset Shape:", clean_df.shape)
print("🔍 Raw Dataset Shape:", raw_df.shape)

# --- Column Preview ---
print("\n🧱 Columns in Cleaned Dataset:\n", clean_df.columns.tolist())
print("\n🧱 Columns in Raw Dataset:\n", raw_df.columns.tolist())

# --- Display First 5 Rows ---
print("\n🔎 Cleaned Dataset Sample:")
display(clean_df.head())

print("\n🔎 Raw Dataset Sample:")
display(raw_df.head())

# --- Data Types and Nulls ---
print("\n🛠 Cleaned Dataset Info:")
display(clean_df.info())

print("\n🛠 Raw Dataset Info:")
display(raw_df.info())

print("\n🧼 Missing Values in Cleaned Dataset:")
display(clean_df.isnull().sum())

print("\n🧼 Missing Values in Raw Dataset:")
display(raw_df.isnull().sum())

# --- Basic Stats ---
print("\n📊 Cleaned Dataset Description:")
display(clean_df.describe(include='all'))

print("\n📊 Raw Dataset Description:")
display(raw_df.describe(include='all'))

# --- Target Variable Distribution ---
print("\n🎯 Accident Severity (Cleaned Dataset):")
display(clean_df['Accident_severity'].value_counts(normalize=True) * 100)

print("\n🎯 Accident Severity (Raw Dataset):")
display(raw_df['Accident_severity'].value_counts(normalize=True) * 100)

# --- Optional: Compare unique values in key columns ---
common_columns = list(set(clean_df.columns) & set(raw_df.columns))
print(f"\n🔁 Common Columns ({len(common_columns)}):\n", common_columns)

for col in common_columns:
    print(f"\n🔹 Column: {col}")
    print("  Unique (cleaned):", clean_df[col].nunique())
    print("  Unique (raw):    ", raw_df[col].nunique())
```
