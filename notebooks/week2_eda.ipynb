This notebook performs an exploratory analysis on the cleaned road accident dataset, focusing on patterns related to accident severity, driver demographics, road and weather conditions.



1. Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('cleaned.csv')
df.head()
```

2. Check for Missing Values

```python
missing = df.isnull().mean().sort_values(ascending=False) * 100
missing = missing[missing > 0]

if not missing.empty:
    plt.figure(figsize=(10,6))
    missing.plot(kind='barh', color='skyblue')
    plt.title('Missing Values Percentage by Feature')
    plt.xlabel('% Missing')
    plt.tight_layout()
    plt.show()
else:
    print("✅ No missing values found in the dataset!")
```

3. Accident Severity Distribution

```python
sns.countplot(x='Accident_severity', data=df, palette='viridis')
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()
```

4. Driver Demographics

4.1 Age Band of Driver

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Age_band_of_driver', data=df, hue='Accident_severity', palette='magma')
plt.title('Accident Severity by Age Band of Driver')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4.2 Sex of Driver

```python
sns.countplot(x='Sex_of_driver', data=df, hue='Accident_severity', palette='Set2')
plt.title('Accident Severity by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

5. Road and Environment Conditions

5.1 Light Conditions

```python
sns.countplot(x='Light_conditions', data=df, hue='Accident_severity', palette='cubehelix')
plt.title('Accident Severity by Light Conditions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5.2 Weather Conditions

```python
sns.countplot(x='Weather_conditions', data=df, hue='Accident_severity', palette='coolwarm')
plt.title('Accident Severity by Weather Conditions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5.3 Type of Collision

```python
plt.figure(figsize=(10,4))
sns.countplot(x='Type_of_collision', data=df, hue='Accident_severity', palette='YlOrBr')
plt.title('Accident Severity by Type of Collision')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Key Insights
- Most accidents fall under a specific severity level, based on your dataset's distribution.
- Younger and less experienced drivers may be more prone to severe accidents.
- Light and weather conditions show visible patterns with accident severity.
- Gender and educational level offer key demographics for prevention focus.
